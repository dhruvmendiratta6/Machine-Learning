```python
import gensim
from gensim.models import word2vec
from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
```

```python
keyed_vectors = gensim.models.KeyedVectors.load_word2vec_format('../Data Sets/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
cosine_similarity([keyed_vectors['apple']], [keyed_vectors['mango']])
```

```python
def odd_one(words):
    all = [keyed_vectors[w] for w in words]
    avg = np.mean(all, axis=0)
    
    min = 1.0
    odd = None
    for w in words:
        sim = cosine_similarity([keyed_vectors[w]], [avg])
        if sim < min:
            min = sim
            odd = w
    
    return odd
```

```python
odd_one(['apple', 'mango', 'party'])
```
