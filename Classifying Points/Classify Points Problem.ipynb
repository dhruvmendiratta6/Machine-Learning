```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train = pd.read_csv(r'../Data Sets/Classify Points Problem/Logistic_X_Train.csv').values
y_train = pd.read_csv(r'../Data Sets/Classify Points Problem/Logistic_Y_Train.csv')
X_test = pd.read_csv(r'../Data Sets/Classify Points Problem/Logistic_X_Test.csv').values
```

```python
X_train.shape
```

```python
y_train
```

```python
ohe = OneHotEncoder(categories='auto')
```

```python
y_hot = ohe.fit_transform(y_train.reshape(-1,1)).todense()
```

```python
y_hot
```

```python
model = Sequential()
```

```python
model.add(Dense(16, activation='relu', input_shape=[2]))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_hot, epochs=40, batch_size=100)
```

```python
out = model.predict(X_test)
```

```python
out
```

```python
y_pred = np.argmax(out, axis=1)
y_pred = y_pred.reshape(-1, 1)
```

```python
df = pd.DataFrame({'label':y_pred[:, 0]})
```
