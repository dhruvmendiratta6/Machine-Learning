```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=100, n_features=10,centers=2, random_state=42)
```

```python
# make_blobs?
```

```python
X.shape
```

```python
# PCA?
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
pca.components_
```

```python
X1 = pca.transform(X)
```

```python
plt.scatter(X1[:, 0], X1[:, 1], c=y)
plt.ylim(-15,15)
```

```python
X_dash = np.dot(X, pca.components_.T)
```

```python
X_dash.shape
```

```python
plt.scatter(X_dash[:, 0], X_dash[:, 1], c=y)
plt.ylim(-15,15)
```

```python
plt.scatter(X_dash[:,0], np.ones(X_dash.shape[0]), c=y)
```

```python
plt.scatter(np.ones(X_dash.shape[0]), X_dash[:,1], c=y)
```
