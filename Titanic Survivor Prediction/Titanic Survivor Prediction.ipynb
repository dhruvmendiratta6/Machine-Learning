```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
```

```python
data_train = pd.read_csv('../data sets/Titanic Survivor Prediction/Train.csv')
data_test = pd.read_csv('../data sets/Titanic Survivor Prediction/Test.csv')
```

```python
data_test.head()
```

```python
data_train.head()
```

```python
data_train.columns
```

```python
selected = ['pclass', 'survived', 'sex', 'age', 'sibsp', 'parch']
```

```python
X = data_train.loc[:,selected]
X_test = data_test.loc[:,selected]
```

```python
X_test.head()
```

```python
X.age.fillna(np.mean(X.age), inplace=True)
X_test.age.fillna(np.mean(X.age), inplace=True)
```

```python
X_test.head()
```

```python
le = LabelEncoder()
```

```python
X.sex = le.fit_transform(X.sex)
X_test.sex = le.fit_transform(X_test.sex)
```

```python
X.head()
```

```python
X_test.head()
```

```python
y_train = X['survived']
X_train = X[['pclass', 'sex', 'age', 'sibsp', 'parch']]
X_test = X_test[['pclass', 'sex', 'age', 'sibsp', 'parch']]
```

```python
X_test.head()
```

```python
model = RandomForestClassifier(n_estimators=20, max_depth=12, criterion='entropy')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```
