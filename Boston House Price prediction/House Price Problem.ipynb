```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv(r'..\Data Sets\House Price Problem\Train_Data.csv')
df1 = pd.read_csv(r'..\Data Sets\House Price Problem\Test_Data.csv')
```

```python
df.head()
df1.head()
```

```python
df.fillna('None', inplace=True)
df = df.drop(['Id'], axis=1)
df1.fillna('None', inplace=True)
df1 = df1.drop(['Id'], axis=1)
```

```python
df.head()
```

```python
df1.head()
```

```python
le = LabelEncoder()
```

```python
df_temp = df.astype('str').apply(le.fit_transform)
df1_temp = df1.astype('str').apply(le.fit_transform)
```

```python
df_temp.head()
```

```python
df1_temp.head()
```

```python
le.classes_
```

```python
X = df_temp.values
X_test = df1_temp.values
```

```python
X_test.shape
```

```python
X = X[:, :-1]
```

```python
X.shape
```

```python
y = df['SalePrice'].values
```

```python
np.unique(y).shape
```

```python
y = y.reshape(-1, 1)
```

```python
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()


print(scaler_x.fit(X))
X_train = scaler_x.transform(X)
X_test = scaler_x.transform(X_test)

print(scaler_y.fit(y))
y_train = scaler_y.transform(y)
```

```python
X_test
```

# Put this in Nueral Network

```python
model = Sequential()
```

```python
model.add(Dense(128, activation='relu',kernel_initializer='normal' , input_shape=[79]))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))

model.summary()
```

```python
model.compile(optimizer= Adam(lr=0.0005), loss='mse', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=1000)
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
pred = scaler_y.inverse_transform(pred)
```
