```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
img=cv2.imread("pan.jpg")
```

```python
# cv2.imshow("image",img)
```

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
X=rgb_img.reshape((-1,3))
X.shape
```

```python
model=KMeans(4)
model.fit(X)
```

```python
color_pellet=model.cluster_centers_
# color_pellet=color_pellet.astype(int)
color_pellet
```

```python
color_pellet.shape[0]
```

```python
for i in range(color_pellet.shape[0]):
    color_img=np.zeros((100,100,3))
    color_img[:,:,:]=color_pellet[i]
#     color_img.astype(int)
    plt.figure()
    plt.imshow(color_img.astype(int))
```

```python
gen=np.zeros(X.shape)
print(gen.shape)
for i in range(len(gen)):
    gen[i]=color_pellet[model.labels_[i]]
    
gen=gen.reshape(rgb_img.shape)
print(gen.shape)
```

```python
plt.figure()
plt.imshow(gen.astype(int))
```
